## Q1 (1 point) — Answer below

```python
import pandas as pd

FILE_PATH = "/content/support_messages.txt"
df = pd.read_csv(FILE_PATH, sep="\t", names=["id", "message"])

print("Shape:", df.shape)
print(df.head(3))
```

## Q2 (3 points) — Answer below

```python
# Q2 — ANSWER CELL

# Word count
df["word_count"] = df["message"].str.split().apply(len)

# Character count
df["char_count"] = df["message"].str.len()

# Number of digits
df["num_count"] = df["message"].str.count(r"\d")

# ALL-CAPS word count
df["upper_word_count"] = df["message"].apply(
    lambda x: sum(1 for word in str(x).split() if word.isupper())
)

# Display full DataFrame
print(df)
```

## Q3 (3 points) — Answer below

```python
# Q3 — ANSWER CELL

import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

STOPWORDS = set(ENGLISH_STOP_WORDS)

def clean_text(text: str) -> str:
    # 1. Lowercase
    text = str(text).lower()

    # 2. Remove punctuation/symbols (keep letters/numbers/spaces)
    text = re.sub(r"[^a-z0-9\s]", "", text)

    # 3. Remove stopwords
    tokens = text.split()
    tokens = [word for word in tokens if word not in STOPWORDS]

    # 4. Remove extra spaces
    text = " ".join(tokens)
    text = re.sub(r"\s+", " ", text).strip()

    return text

# Create clean column
df["clean"] = df["message"].apply(clean_text)

# Print original and clean for id=1 and id=4
print("ID = 1")
print("Original:", df.loc[df["id"] == '1', "message"].values[0])
print("Clean:", df.loc[df["id"] == '1', "clean"].values[0])

print("\nID = 4")
print("Original:", df.loc[df["id"] == '4', "message"].values[0])
print("Clean:", df.loc[df["id"] == '4', "clean"].values[0])
```

## Q4 (2 points) — Answer below

```python
# Q4 — ANSWER CELL

# Extract order_id
df["order_id"] = df["message"].str.extract(
    r"(ORD-\d{4})", flags=re.IGNORECASE
)

# Extract email
df["email"] = df["message"].str.extract(
    r"([A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,})"
)

# Display required columns
print(df[["id", "order_id", "email"]])
```

## Q5 (1 point) — Answer below

```python
# Q5 — ANSWER CELL

from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Fit TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df["clean"])

# Compute average TF-IDF score per term
avg_tfidf = np.asarray(tfidf_matrix.mean(axis=0)).ravel()

# Get feature names
terms = vectorizer.get_feature_names_out()

# Create DataFrame of terms and scores
tfidf_df = pd.DataFrame({
    "term": terms,
    "avg_tfidf": avg_tfidf
})

# Sort and get top 5
top5 = tfidf_df.sort_values(by="avg_tfidf", ascending=False).head(5)

print("Top 5 TF-IDF Keywords:")
print(top5)
```
